# Symmetrize

```python
import mnx
```

```python
structure = mnx.Structure.from_file("./structs/CsV3Sb5/CONTCAR_2x2x2_5gpa", format="vasp")
```

```python
structure.info()
```

```python
prim_structure = structure.to_primitive()
```

```python
prim_structure.info()
```

```python
sym_structure = structure.to_primitive(1e0, rotate=False)
```

```python
sym_structure.info()
```

```python
%matplotlib widget
```

```python
#Recommended to use an interactive backend!

kpoints = [[0.5,0,0], [0.5,0,0.5], [0,0,0.5]]
ax = sym_structure.plot_bz(color="blue", kpoints=kpoints, labels=["M","L","A"])
kpoints = [[1,0.5,0.5], [0.5,0,0.5], [0.5,0.5,0]]
ax = prim_structure.plot_bz(ax, color="red", kpoints=kpoints, labels=["T","Y","Z"])
```
